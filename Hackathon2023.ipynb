```python
import pandas as pd
from datetime import datetime

# define a function to convert a date string to the day of the week
def get_day_of_week(date_str):
    date_obj = datetime.strptime(date_str, '%d/%m/%Y')
    return date_obj.weekday()

# create a new column 'day_of_week' by applying the get_day_of_week function to the 'date' column
df['day_of_week'] = df['date'].apply(lambda x: get_day_of_week(x))

# map the integer values of 'day_of_week' to day of the week names
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week_name'] = df['day_of_week'].map(lambda x: day_names[x])

```

```python

# define a function to check if the value is 6 or 7 and return 1 or 0
def is_weekend(day):
    if day in [1, 7]:
        return 1
    else:
        return 0

# apply the function to the 'day_of_week' column and store the result in a new column 'Weekend'
df['Weekend'] = df['day_of_week'].apply(is_weekend)
```
